# < Numpy >

print(type()) : 인자의 class 확인

print(ndarray.dtype) : ndarray의 데이터 타입 확인

print(ndarray.ndim) : ndarray의 차원 확인

print(ndarray.shape) : ndarray의 차원과 요소 수를 tuple로 알려줌

ndarray.astype(np.int32) : ndarray를 숫자형 데이터 타입으로 바꿈

ndarray.astype(np.str_) : ndarray를 문자형 데이터 타입으로 바꿈

---

np.array() : 인자(list)로 ndarray 만들기

np.zeros() : 주어진 shape에 대한 ndarray를 생성하고 0으로 채움

np.ones() : 주어진 shape에 대한 ndarray를 생성하고 1로 채움

np.empty() : 주어진 shape에 대한 ndarray를 생성하고 garbage 값으로 채움

np.full((2,3), 7*, dtype=np.float64*) : # 2행 3열 ndarray를 생성하고 실수 7로 채움

np.zeros_like(ndarray) : ndarray의 shape과 같은 형태로 ndarray를 생성하고 0으로 채움. 인자로 들어간 데이터 타입을 따름

np.arange() : Numpy의 arange == Python의 range

np.arange(1.3, 10.1*, 2*) : 1.3부터 10.1까지 2씩 증가하는 값으로 구성된 ndarray

np.linspace(start, stop, num) : 시작과 끝을 모두 포함, num은 그 안에 균일한 간격으로 몇 개의 숫자가 들어갈지. (start - stop) / (num - 1). 균일한 간격의 데이터를 생성해서 ndarray를 만들어냄

---

1. np.random.normal(mean, std, shape) : ndarray를 생성하고 랜덤한 실수형 난수로 채움. 정규분포. mean·std·shape가 옵션

2. np.random.rand() : ndarray를 생성하고 랜덤한 실수형 난수로 채움. 균등분포. [0, 1) 0부터 1 사이(0은 포함, 1은 불포함). 인자가 하나면 일 차원

3. np.random.randn() : ndarray를 생성하고 랜덤한 실수형 난수로 채움. 표준 정규분포. 평균이 0이고 표준편차가 1

4. np.random.randint(low, high, shape) : ndarray를 생성하고 랜덤한 정수형 난수로 채움. 균등분포. low·high·shape가 옵션

5. np.random.random() : ndarray를 생성하고 랜덤한 실수형 난수로 채움. 균등분포. [0, 1) 0부터 1사이(0은 포함, 1은 불포함). rand()와 달리 shape가 옵션

np.random.seed() : 랜덤의 초기 값을 고정하면 난수의 재현성이 확보됨. None은 난수

np.random.shuffle(ndarray) : 원본 data의 순서를 바꿈. 초기 값 고정 가능

np.random.choice(a, size, replace=True/False, p=[총합은 1. 실수]) : sampling에 사용. a는 모집단이자 배열, size는 숫자 몇 개를 뽑을 건지, replace가 True이면 뽑았던 것 다시 뽑을 수 있음, p는 각 데이터가 뽑힐 수 있는 확률

plt.hist(ndarray, bins=100) : 히스토그램(도수표)을 생성. bins는 간격

plt.show() : 그래프 출력

---

ndarray.reshape() : shape을 바꿔서 새로운 ndarray를 만드는 것이 아니고, 원래 ndarray의 데이터를 다르게 보여주는 view를 생성. 요소 개수가 맞지 않을 경우 에러

ndarray.ravel() : 무조건 1차원의 view를 만듦

ndarray.resize() : view를 만들지 않고 원본을 바꿈. 요소 개수가 적으면 추가, 많으면 버려서 데이터가 유실됨

---

print(ndarray[행,열]) : ndarray의 indexing

print(ndarray[행:, 열:]) : ndarray의 slicing

my_mask = (ndarray % 2 == 0) : Boolean indexing

print(ndarray[my_mask])

print((ndarray > 10).sum()) : 10보다 큰 수는 몇 개? True = 1

print(ndarray[0:2, [1,3]]) : Fancy Indexing. 행에 대한 slicing, 열에 대한 indexing

print(ndarray[np.ix_([0,2], [1,3])]) : Fancy Indexing

my_iter.iternext() : iterator 다음으로 이동

---

print(ndarray[idx], idx)

my_iter.iternext()

np.matmul(ndarray1, ndarray2) : 행렬곱 연산. 앞의 2차원 matrix 열과 뒤의 2차원 matrix 행 개수가 같아야 함

ndarray.T : 전치행렬(transpose)은 행은 열로, 열은 행으로 바꿔줌

print(np.sum(ndarray)) : 모든 요소의 합. Numpy가 제공하는 함수

print(ndarray.sum()) : 모든 요소의 합. ndarray가 가진 메서드

print(ndarray.sum(axis=0)) : 행 방향 axis를 부여해서 열끼리 합을 구함

print(ndarray.sum(axis=1)) : 열 방향 axis를 부여해서 행끼리 합을 구함

print(ndarray.mean()) : 평균

print(ndarray.max()) : 최대값

print(ndarray.min()) : 최소값

print(ndarray.argmax()) : 가장 큰 값의 인덱스

print(ndarray.std()) : 표준편차
