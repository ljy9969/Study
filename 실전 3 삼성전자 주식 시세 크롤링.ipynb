## 하루 Ver.

```python
import requests
from bs4 import BeautifulSoup

URL = 'https://finance.naver.com/item/sise_day.naver?code=005930&page=1'
headers = {'user-agent': 'Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:103.0) Gecko/20100101 Firefox/103.0'}

res = requests.get(URL, headers = headers)
res.status_code
soup = BeautifulSoup(res.text, 'html.parser')
infos = soup.select('span.tah')
print(f'element {len(infos)}개')

results = []
keys = ['날짜', '종가', '전일비', '시가', '고가', '저가', '거래량']
values = []
volume_sum = 0

for i in range(70):    
    a = infos[i].text.strip()
    values.append(a)
    
    if len(values) % 7 == 0:
        info_dict = dict(zip(keys, values))
        results.append(info_dict)
        volume_sum += int(values[-1].replace(',', ''))
        values = []
    else:
        continue
    
print(f'{len(results)}일 동안의 거래량 총액 ' + format(volume_sum, ',d') +'원')
results
```

## 10 페이지 Ver.

```python
import requests
from bs4 import BeautifulSoup

headers = {'user-agent': 'Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:103.0) Gecko/20100101 Firefox/103.0'}
final_results = []

for page in range(1, 11):
    URL = f'https://finance.naver.com/item/sise_day.naver?code=005930&page={page}'
    res = requests.get(URL, headers = headers)
    soup = BeautifulSoup(res.text, 'html.parser')
    infos = soup.select('span.tah')

    results = []
    keys = ['날짜', '종가', '전일비', '시가', '고가', '저가', '거래량']
    values = []

    volume_sum = 0

    for i in range(70):    
        a = infos[i].text.strip()
        values.append(a)

        if len(values) % 7 == 0:
            info_dict = dict(zip(keys, values))
            results.append(info_dict)
            volume_sum += int(values[-1].replace(',', ''))
            values = []
        else:
            continue
    final_results.append(results)
    print(f'{len(results)}일 동안의 거래량 총액 ' + format(volume_sum, ',d') +'원')
    
final_results
```
